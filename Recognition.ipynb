```python
import cv2
import numpy as np
import pickle
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import VGG16

# Load the face classifier and label encoder
with open('face_classifier.pkl', 'rb') as file:
    face_classifier, face_label_encoder = pickle.load(file)
    
# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the VGG16 model pre-trained on ImageNet
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

def preprocess_image(img_array):
    img = cv2.resize(img_array, (224, 224))
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img

def extract_feature(img_array):
    img = preprocess_image(img_array)
    feature = vgg_model.predict(img).flatten()
    return feature

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        face_crop = frame[y:y+h, x:x+w]
        feature = extract_feature(face_crop)

        # Predict the person
        face_prediction = face_classifier.predict([feature])
        face_name = face_label_encoder.inverse_transform(face_prediction)[0]

        # Display the results
        label = f'{face_name}'
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    cv2.imshow('Face and Beard Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
